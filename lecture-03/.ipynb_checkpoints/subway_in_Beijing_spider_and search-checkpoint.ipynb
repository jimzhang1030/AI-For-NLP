# 爬取北京地铁站信息

```python
import re
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://baike.baidu.com/item/%E5%8C%97%E4%BA%AC%E5%9C%B0%E9%93%81/408485'
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36'
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content.decode('utf8'), 'lxml') # 也可写'html_parse
```

```python
from collections import defaultdict
```

```python
page_list = defaultdict(list)
for i in soup.find_all('a', {'target':'_blank'}):
    temp = re.findall('^北京地铁[\w|\d]+线', str(i.string))
    if temp != []:
        if str(i.string) not in page_list.keys():
            page_list[str(i.string)] = str(i['href'])
        else:
            continue
```

```python
page_list
```

```python
for i in page_list.keys():
    page_list[i] = 'https://baike.baidu.com' + page_list[i]
```

```python
f_stations = []
f_distances = []
# len_s = len(f_stations)
# len_d = len(f_distances)
nodes_line = defaultdict(set)
for i in page_list.keys():
    tmp_url = page_list[i]
    tmp_response = requests.get(tmp_url, headers=headers)
    tmp_soup = BeautifulSoup(tmp_response.content.decode('utf8'), 'lxml')
    
    print('now is in{}'.format(i))
    
    if i == '北京地铁16号线':
        for j in tmp_soup.find_all('b', text='站区间'):
            for k in j.parent.parent.find_next_siblings():
                if k.contents[0].string:
#                     print()
                    f_stations += re.findall('(.*?)～(.*)', k.contents[0].string)
                    if re.findall('(.*?)～(.*)', k.contents[0].string)!= []:
                        nodes_line[re.findall('(.*?)～(.*)', k.contents[0].string)[0][0]].add(i)
                        nodes_line[re.findall('(.*?)～(.*)', k.contents[0].string)[0][1]].add(i)
                    f_stations += re.findall('(.*?)~(.*)', k.contents[0].string)
                    if re.findall('(.*?)~(.*)', k.contents[0].string)!=[]:
                        nodes_line[re.findall('(.*?)~(.*)', k.contents[0].string)[0][0]].add(i)
                        nodes_line[re.findall('(.*?)~(.*)', k.contents[0].string)[0][1]].add(i)
                    f_distances += [float(k.contents[1].string) * 1000]
        continue
    for j in tmp_soup.find_all('th', text='起始/终到车站'):
        for k in j.parent.find_next_siblings():
            if k.contents[0].string:
                f_stations += re.findall('(.*?)——(.*)', k.contents[0].string)
#                 print(re.findall('(.*?)——(.*)', k.contents[0].string))
                nodes_line[re.findall('(.*?)——(.*)', k.contents[0].string)[0][0]].add(i)
                nodes_line[re.findall('(.*?)——(.*)', k.contents[0].string)[0][1]].add(i)
                if k.contents[1].string[-2] == '千':
                    f_distances += [float(k.contents[1].string[:-2]) * 1000]
                else:
                    f_distances += re.findall('\d+', k.contents[1].string)
    print(len(f_distances), len(f_stations))
        
# for i, v in f_stations:
#     f_stations_connections[i] = v
#     print(f_stations)
#     print(f_distances)
#     len_s = len(f_stations)
#     len_d = len(f_distances)
#     ic(len_s)
#     ic(len_d)
```

```python
def is_goal(destination):
    def _wrap(current_path):
        return current_path[-1] == destination
    return _wrap
```

```python
is_goal('兰州')
```

```python
#最短路程
def get_path_distance(path):
    distance = 0
    for i, s in enumerate(path[:-1]):
#         ic(s)
#         print(distances_dict[(s,path[i + 1])])
        distance += int(distances_dict[(s,path[i + 1])])
    return distance
```

```python
#最小换乘
def get_total_swaps(path):
    count = 0
    if len(path) > 2:
        for i, s in enumerate(path[:-1]):
            exchange = 1
            for j in nodes_line[s]:
                exchange = 0 if j in nodes_line[path[i+1]] else 1
            if exchange == 0:
                count += 1
        return count
    else: return 0
```

```python
def sort_path(cmp_func, beam=-1):
    def _sorted(pathes):
        return sorted(pathes, key=cmp_func)[:beam]
    return _sorted
```

```python
def search(graph, start, is_goal, search_strategy):
    pathes = [[start]]
    seen = set()
#     seen.add(start)
    while pathes:

        path = pathes.pop(0)
        
        frontier = path[-1]
    
        if frontier in seen: continue
    
        successors = f_stations_connections[path[-1]]
    
        for i in successors:
            if i in path: continue
            pathes.append(path + [i])
            if is_goal(path + [i]):
                return path + [i]
        
        seen.add(frontier)
#         ic(pathes)
        pathes = search_strategy(pathes)
#         ic(pathes)
        
        
```

```python
#最短距离
search(f_stations_connections, '苹果园', is_goal=is_goal('五道口'), search_strategy=sort_path(get_path_distance, beam=20))
```

```python
#最小换乘
search(f_stations_connections, '苹果园', is_goal=is_goal('五道口'), search_strategy=sort_path(get_total_swaps, beam=20))
```

```python
def get_comprehensive_path(path):
    return get_path_distance(path) + 1000000 * get_total_swaps(path)
```

```python
#综合
search(f_stations_connections, '苹果园', is_goal=is_goal('五道口'), search_strategy=sort_path(get_comprehensive_path, beam=20))
```
